# Convert sklearn.utils.Bunch into Pandas DataFrame
Find more on- https://github.com/viswanathanc/basic_python/blob/master/sklearn.utils.bunch%20to%20pandas%20Dataframe.ipynb

```python
data = pd.DataFrame(data = iris.data, columns = iris.feature_names)
data['species'] = iris.target
data.species = data.species.apply(lambda x: iris.target_names[x])
data.head(10)
```

```python
print(data.info())
print('----------------------------------------------------------')
print(data.groupby('species').size())
print('----------------------------------------------------------')

data.describe()
```

# Now Split the Dataset
We should split the dataset before applying any standardization/normalization or visulaization to prevent data leakage.

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.3, random_state = 5)
train.shape, test.shape
```

#Visualization

```python
# Histogram
fig, ax = plt.subplots(2, 2, figsize = (8,5))
sns.histplot(data = train, x = 'sepal length (cm)', ax = ax[0,0], bins = 20, kde = True)
ax[0,0].set_title('Sepal Length (cm)')
sns.histplot(data = train, x = 'sepal width (cm)', ax = ax[0,1], bins = 20, kde = True)
ax[0,1].set_title('Sepal Width (cm)')
sns.histplot(data = train, x = 'petal length (cm)', ax = ax[1,0], bins = 20, kde = True)
ax[1,0].set_title('Petal Length (cm)')
sns.histplot(data = train, x = 'petal width (cm)', ax = ax[1,1], bins = 20, kde = True)
ax[1,1].set_title('Petal Width (cm)')
fig.tight_layout()
plt.show()
```

From pairplot we clearly see that we can't classify species properly with sepal length and sepal width. We have to choose petal length and petal width.

```python
corrmat = train.corr()
top_corr_features = corrmat.index
g=sns.heatmap(data[top_corr_features].corr(), annot=True, square = True, cmap="RdYlGn")
plt.xticks(rotation=45)
plt.show();
```

```python
X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]
X_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1]
X_test.shape, y_test.shape
```

# K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier  
from sklearn import metrics

neighbors = np.arange(1,9)
train_accuracy = np.zeros(len(neighbors))
test_accuracy = np.zeros(len(neighbors))

#Fit dataset in K-NN model
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    #Fit the model
    knn.fit(X_train, y_train)
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train,y_train)
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
#Plotting the testing and training accuracy for different n_neighbours
plt.figure(figsize=(7,5))
plt.title('KNN Accuracy with Varying Numbers of Neighbors',fontsize=15)
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend(prop={'size': 15})
plt.xlabel('Number of neighbors',fontsize=15)
plt.ylabel('Accuracy',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.show()
```

```python
#From the graph we see that the training and test accuracy is highiest for k = 7.
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
train_accuracy = knn.score(X_train, y_train)
test_accuracy = knn.score(X_test, y_test)
print("For K=7, train accuracy =", train_accuracy, " and test accuracy =", test_accuracy)
y_pred = knn.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Prediction accuracy =", accuracy)
```

```python
metrics.plot_confusion_matrix(knn, X_test, y_test, cmap=plt.cm.Blues)
```

Wee see 1 versicolor iris flower was labeled as virginica. Means 1 false positive for virginica. This will lower viirginica precision. And we couldn't detect one versicolor (true positive missing). Thus versicolor recall will decrease.

# Encoding target labels

```python
from sklearn.preprocessing import LabelEncoder
labels = data.species.to_numpy()
y = LabelEncoder().fit_transform(labels)
print(y)
X = data.drop(columns=['sepal length (cm)', 'sepal width (cm)','species'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier(max_depth = 3, random_state = 5)
dt.fit(X_train, y_train)
print('Train Accuracy: ',dt.score(X_train, y_train))
print('Test Accuracy: ',dt.score(X_test, y_test))
y_pred = dt.predict(X_test)
print('Prediction Accuracy: ',metrics.accuracy_score(y_test, y_pred))

plt.figure(figsize= (10,8))
plot_tree(dt, feature_names = ['petal length (cm)', 'petal width (cm)'], class_names = iris.target_names, filled = True)
plt.show();
```

# Ensemble Method - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier().fit(X_train, y_train)
print('Train Accuracy: ',rfc.score(X_train, y_train))
print('Test Accuracy: ',rfc.score(X_test, y_test))
y_pred = rfc.predict(X_test)
print('Prediction Accuracy: ',metrics.accuracy_score(y_test, y_pred))
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
grid_params = [
               {
                'n_estimators': [20, 50, 100],
                'max_features': [1,2],
                'max_depth': [2,3,5]
               }
]
rfc = RandomForestClassifier(random_state= 5)
grid_search = GridSearchCV(rfc, grid_params, scoring = 'accuracy', cv =3)
grid_search.fit(iris.data,iris.target)
```

```python
print(grid_search.best_params_)
result = pd.DataFrame(grid_search.cv_results_,
                      columns = ['param_max_depth','param_max_features','param_n_estimators','mean_test_score','rank_test_score'])
result.sort_values('rank_test_score', ascending=True, axis=0)
```

```python
from sklearn.svm import SVC
svm = SVC(random_state=5)
grid_params = [
               {
                'kernel':['linear','rbf'],
                'C': [0.1,1,10,100],
                'gamma': [0.01,0.001,0.0001]
               }
]

svmCV = GridSearchCV(svm, grid_params, scoring = 'accuracy', cv =3)
svmCV.fit(iris.data,iris.target)
```

```python
print(svmCV.best_params_)
result = pd.DataFrame(svmCV.cv_results_,
                      columns = ['params','mean_test_score','rank_test_score'])
result.sort_values('rank_test_score', ascending=True, axis=0)
```
